***Are there any duplicated rows?***

```python
df.duplicated().sum()
```

***What are the top 3 years with the most roller coasters introduced?***

```python
g=df.groupby('Year_Introduced').count()
g=g.sort_values(by='Coaster_Name',ascending=[False])
g.iloc[0:3,0]
```

***What is the average speed? Also display a plot to show it’s distribution.***

```python
df['Speed_mph']=df['Speed_mph'].fillna(0)
df['Speed_mph'].mean()
```

```python
bin=range(0,200,7) #any list as input
plt.hist(df.Speed_mph,bins=bin,color='#ffceff')
plt.xticks(bin[:30:2],rotation=True)
plt.xlabel('Speed_mph')
#plt.ylabel('No of passenger')
plt.show()
```
