***What is the frequency of the dataset? (The time period between each row)***

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df['frequency']=df['Date'] - df['Date'].shift()
df['frequency']
```

***What is the data type of the index?
Set the index to a Datetime.***

```python
df.index.dtype
```

```python
df.index=pd.to_datetime(df.index)
```

```python
df.index.dtype
```

***Change the frequency to monthly, sum the values and assign it to new variable called monthly.***

```python
df['month']=df['Date'].dt.month
df['year']=df['Date'].dt.year
```

```python
df.head(2)
```

```python
monthly=df.groupby(['year','month']).sum()
monthly
```

*** Now change the frequency to year and assign to a new variable called year.***

```python
g_year=df.groupby('year').sum()
g_year
```
